# Q1:Race

```python
def race(x, y):
    """The tortoise always walks x feet per minute, while the hare repeatedly
    runs y feet per minute for 5 minutes, then rests for 5 minutes. Return how
    many minutes pass until the tortoise first catches up to the hare.

    >>> race(5, 7)  # After 7 minutes, both have gone 35 steps
    7
    >>> race(2, 4) # After 10 minutes, both have gone 20 steps
    10
    """
    assert y > x and y <= 2 * x, 'the hare must be fast but not too fast'
    tortoise, hare, minutes = 0, 0, 0
    while minutes == 0 or tortoise - hare: 
        tortoise += x
        print(f"tortoise:{tortoise} at {minutes+1}min")# 
        if minutes % 10 < 5:
            hare += y
        print(f"hare:{hare} at {minutes+1}min")#
        minutes += 1

        # test break, when minutes is too large, I think it runs forever.
        if minutes > 1000:
            break

    return minutes
```

Find positive integers $x$ and $y$ with $x\lt y \leq 2x$ for which either 
- race(x,y) returns the wrong value
- race(x,y) runs forever

```python
race(3,4) # runs forever
```

```python
race(2,3) # returns a wrong value. it is not the first time the tortoise passes the hare. Because the code only count some discrete points, miss the correct point.
```
